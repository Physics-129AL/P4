```python
# Problem 3 A

# Steps: (sorry for my half latex integral notations)

# M_x(t)= \int_{0}^{inf} dx * e^(tx) * Z(lambda) * lamda * e^-(lambda*x)

# = \int_{0}^{inf} dx * Z(lambda) * lamda * e^[-(lambda-t)*x]

# = Z(lambda) * lambda * \int_{0}^{inf} dx * e^[-(lambda-t)*x]

# = Z(lambda) * lamdba *[-1/(lambda-t) * e^(-(lambda-t)x)]{0}^{inf}

# = Z(lambda) * lambda * [-1/(lambda-t) * (0-1)]

# = Z(lambda) * lambda / (t-lambda)
```

```python
# Problem 3 B

# Steps:
# Take derivatives of MGF w/ respect to t and then evalute at t=0
# MEAN: 

# E[X] = dM_x(t)/dt | t=0

# dM_x(t)/dt = d/dt (Z(lambda)*lambda/(t-lambda))

# = -Z(lambda)*lambda/(t-lambda)^2

# plugging in t=0:

# E[X] = -Z(lambda)*lambda/(-lambda)^2

# = Z(lambda)*lambda/lambda^2

# = Z(lambda)/lambda

# Variance

# Var(X) = d^2/(dt^2)*(M_x(t)) | t=0 + [dM_x(t)/dt | t=0]^2

# d^2 M_x(t)/dt^2 = d/dt*(-Z(lambda)*lambda/(t-lambda)^2)

# = 2*Z(lambda)*lambda/(t-lambda)^3

# plugging in t=0

# Var(X)= 2*Z(lambda)/lambda^2 + Z(lambda)^2/lambda^4
```

```python
# Problem 3 C

import numpy as np
import matplotlib.pyplot as plt

lambda_values = np.linspace(0.1, 5.0, 100)

analytical_mean = 1 / lambda_values
analytical_variance = 2 / lambda_values**2 + 1 / lambda_values**4

delta_t = 1e-5  

numerical_mean = []
numerical_variance = []

for lambda_val in lambda_values:
    mean_derivative = (-1 / ((delta_t - lambda_val)**2) + 1 / (lambda_val**2)) / delta_t
    numerical_mean.append(mean_derivative)

    variance_derivative = (2 / ((delta_t - lambda_val)**3) - 2 / (lambda_val**2)) / delta_t
    numerical_variance.append(variance_derivative)

plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.plot(lambda_values, analytical_mean, label='Analytical Mean')
plt.plot(lambda_values, numerical_mean, label='Numerical Mean', linestyle='--')
plt.xlabel('Lambda')
plt.ylabel('Mean')
plt.title('Mean vs. Lambda')
plt.legend()

plt.subplot(122)
plt.plot(lambda_values, analytical_variance, label='Analytical Variance')
plt.plot(lambda_values, numerical_variance, label='Numerical Variance', linestyle='--')
plt.xlabel('Lambda')
plt.ylabel('Variance')
plt.title('Variance vs. Lambda')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Problem 3 D

# Steps:

# M_X(t) = E[e^{tX}] = sum_{x} e^{tx} * P(X=x)

# P(X=k) = p(1-p)^{k-1} for  k ={1, 2, 3,...}

# M_X(t) &= sum_{k=1}^{inf} e^{tk} * p(1-p)^{k-1}

# = p * sum_{k=1}^{infty} (e^t(1-p))^{k-1} since e^{tk} is a geometric series

# = {p}/{1 - e^t(1-p)} using the sum of an infinite geometric series formula

# P(X=k) is {p}/{1 - e^t(1-p)}.
```

```python
# Problem 3 E

# Steps:

# Mean:
# E[X] = {dM_X(t)}/{dt} at t=0


# {dM_X(t)}/{dt} = {d}/{dt}({p}/{1 - e^t(1-p)})
# = {p(1-p)e^t}/{(1 - e^t(1-p))^2}

# at t = 0:

# E[X] = {p(1-p)e^0}/{(1 - e^0(1-p))^2} = {p(1-p)}/{(1 - (1-p))^2} = \frac{p(1-p)}{p^2} = {1-p}/{p}

# Variance:
# Var(X) = {d^2M_X(t)}/{dt^2} + ({dM_X(t)}/{dt})^2 at t=0


# {d^2M_X(t)}/{dt^2} = {d}/{dt}({p(1-p)e^t}/{(1 - e^t(1-p))^2})
# = {p(1-p)e^t(1 - e^t(1-p))^2 + 2p(1-p)e^t(1-p)e^t}/{(1 - e^t(1-p))^4}
# = {p(1-p)e^t(1 - e^t(1-p))^2 + 2p(1-p)e^{2t}(1-p)}/{(1 - e^t(1-p))^4}


# at t = 0:

# {d^2M_X(t)}/{dt^2} = {p(1-p)e^0(1 - e^0(1-p))^2 + 2p(1-p)e^0(1-p)}/{(1 - e^0(1-p))^4}
# = {p(1-p)(1 - (1-p))^2 + 2p(1-p)(1-p)}/{(1 - (1-p))^4}
# = {p(1-p)p^2 + 2p(1-p)p}/{p^4}
# = {p^2(1-p + 2(1-p))}/{p^4}
# = {3(1-p)}/{p^2}


# Also, E[X] = {1-p}/{p}
# Therefore, ({dM_X(t)}/{dt})^2 = ({p(1-p)e^0}/{(1 - e^0(1-p))^2})^2 =({1-p}/{p^2})^2 = {(1-p)^2}/{p^4}

# Var(X) = {3(1-p)}/{p^2} + {(1-p)^2}/{p^4}
```

```python
# Problem 3 F

import numpy as np
import matplotlib.pyplot as plt

p_values = np.linspace(0.1, 0.9, 100)

analytical_mean = (1 - p_values) / p_values
analytical_variance = 3 * (1 - p_values) / (p_values**2) + (1 - p_values)**2 / (p_values**4)

delta_p = 1e-5

numerical_mean = []
numerical_variance = []

for p_val in p_values:
    mean_derivative = ((1 - (p_val + delta_p)) / (p_val + delta_p) - (1 - p_val) / p_val) / delta_p
    numerical_mean.append(mean_derivative)

    variance_derivative = ((3 * (1 - (p_val + delta_p)) / ((p_val + delta_p)**2)) + ((1 - (p_val + delta_p))**2) / ((p_val + delta_p)**4) 
                           - (3 * (1 - p_val) / (p_val**2)) - ((1 - p_val)**2) / (p_val**4)) / delta_p
    numerical_variance.append(variance_derivative)

plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.plot(p_values, analytical_mean, label='Analytical Mean')
plt.plot(p_values, numerical_mean, label='Numerical Mean', linestyle='--')
plt.xlabel('p')
plt.ylabel('Mean')
plt.title('Mean vs. p')
plt.legend()

plt.subplot(122)
plt.plot(p_values, analytical_variance, label='Analytical Variance')
plt.plot(p_values, numerical_variance, label='Numerical Variance', linestyle='--')
plt.xlabel('p')
plt.ylabel('Variance')
plt.title('Variance vs. p')
plt.legend()

plt.tight_layout()
plt.show()
```
