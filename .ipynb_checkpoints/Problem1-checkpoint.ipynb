```python
#Problem 1 A
# load dataset 1
import json

json_file_path = 'problem_1/dataset_1.json'

with open(json_file_path, 'r') as json_file:
    boolean_list_out_1 = json.load(json_file)
```

```python
import matplotlib
import scipy

from scipy.stats import binom

def likelihood(p, M, N):
    return binom.pmf(M, N, p)
```

```python
from scipy.stats import beta

prior_alpha = 1
prior_beta = 1

def prior(p):
    return beta.pdf(p, prior_alpha, prior_beta)
```

```python
def posterior(p, M, N):
    return likelihood(p, M, N) * prior(p)
```

```python
import matplotlib.pyplot as plt
import numpy as np

N = len(boolean_list_out_1)
M = sum(boolean_list_out_1)

p_values = np.linspace(0, 1, 1000)

likelihood_values = likelihood(p_values, M, N)
prior_values = prior(p_values)
posterior_values = posterior(p_values, M, N)

expectation = np.trapz(p_values * posterior_values, p_values)
variance = np.trapz((p_values - expectation)**2 * posterior_values, p_values)

plt.subplot(3, 1, 1)
plt.plot(p_values, posterior_values, label='Dataset 1')
plt.xlabel('p (Probability of Heads)')
plt.ylabel('Posterior Probability')
plt.title('Posterior Distribution for Dataset 1')
plt.legend()

print(f'Expectation for dataset 1: {expectation:.8f}')
print(f'Variance for dataset 1: {variance:.8f}')

# Load dataset 2
json_file_path = 'problem_1/dataset_2.json'
with open(json_file_path, 'r') as json_file:
    boolean_list_out_2 = json.load(json_file)

import matplotlib.pyplot as plt
import numpy as np

N = len(boolean_list_out_2)
M = sum(boolean_list_out_2)

p_values = np.linspace(0, 1, 1000)

likelihood_values = likelihood(p_values, M, N)
prior_values = prior(p_values)
posterior_values = posterior(p_values, M, N)

expectation = np.trapz(p_values * posterior_values, p_values)
variance = np.trapz((p_values - expectation)**2 * posterior_values, p_values)

plt.subplot(3, 1, 2)
plt.plot(p_values, posterior_values, color='r', label='Dataset 2')
plt.xlabel('p (Probability of Heads)')
plt.ylabel('Posterior Probability')
plt.title('Posterior Distribution for Dataset 2')
plt.legend()

print(f'Expectation for dataset 2: {expectation:.8f}')
print(f'Variance for dataset 2: {variance:.8f}')

# Load dataset 3
json_file_path = 'problem_1/dataset_3.json'
with open(json_file_path, 'r') as json_file:
    boolean_list_out_3 = json.load(json_file)

import matplotlib.pyplot as plt
import numpy as np

N = len(boolean_list_out_3)
M = sum(boolean_list_out_3)

p_values = np.linspace(0, 1, 1000)

likelihood_values = likelihood(p_values, M, N)
prior_values = prior(p_values)
posterior_values = posterior(p_values, M, N)

expectation = np.trapz(p_values * posterior_values, p_values)
variance = np.trapz((p_values - expectation)**2 * posterior_values, p_values)

plt.subplot(3, 1, 3)
plt.plot(p_values, posterior_values, color='g', label='Dataset 3')
plt.xlabel('p (Probability of Heads)')
plt.ylabel('Posterior Probability')
plt.title('Posterior Distribution for Dataset 3')
plt.legend()
plt.tight_layout()


print(f'Expectation for dataset 3: {expectation:.8f}')
print(f'Variance for dataset 3: {variance:.8f}')
```

```python
# Problem 1 B

import matplotlib.pyplot as plt
import numpy as np
from scipy.special import gamma, gammaln

def factorial_stirling(n):
    return np.sqrt(2 * np.pi * n) * (n / np.e)**n

def factorial_gamma(n):
    return gamma(n + 1)

N_values = np.linspace(1, 10,1000)

# calculate factorial
factorial_stirling_values = factorial_stirling(N_values)
factorial_gamma_values = factorial_gamma(N_values)

# difference between Stirling approximation and Gamma function
difference_values = factorial_stirling_values - factorial_gamma_values

plt.subplot(2, 1, 1)
plt.plot(N_values, factorial_stirling_values, label="Stirling's Approximation")
plt.plot(N_values, factorial_gamma_values, label="Gamma Function", color='r', linestyle="dotted")
plt.title('Factorial Comparison')
plt.xlabel('N')
plt.ylabel('Factorial Value')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(N_values, difference_values, label="Difference", color="g")
plt.title('Difference between Stirling and Gamma')
plt.xlabel('N')
plt.ylabel('Difference')
plt.legend()

plt.tight_layout()

plt.show()
```

```python
# Problem 1 C

#Steps: 
# P(M,N|p) = (N M) p**M (1-p)**(N-M)
# log(P(M,N|p))= log (N M) + M*log(p)+(N-M)*log(1-p)
# take partial derivative with respect to p
# d/dp(log(P(M,N)|p)) = M/p - (N-M)/(1-p)
# set equal to zero
# M/p = (N-M)/(1-p)
# p = M/N

json_file_paths = ['problem_1/dataset_1.json', 'problem_1/dataset_2.json', 'problem_1/dataset_3.json']

m_n_ratios = []

for json_file_path in json_file_paths:
    with open(json_file_path, 'r') as json_file:
        boolean_list_out = json.load(json_file)
    
    N = len(boolean_list_out)
    M = sum(boolean_list_out)
    
    m_n_ratio = M / N
    
    m_n_ratios.append(m_n_ratio)

for i, ratio in enumerate(m_n_ratios, 1):
    print(f'Data Set {i}: M/N = {ratio:.2f}')
```

```python
# Problem 1 D

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

p_values = np.linspace(0, 1, 1000)

likelihood_values_1 = []
likelihood_values_2 = []
likelihood_values_3 = []

for boolean_list_out in [boolean_list_out_1, boolean_list_out_2, boolean_list_out_3]:
    N = len(boolean_list_out)
    M = sum(boolean_list_out)
    
    likelihood_values = binom.pmf(M, N, p_values)
    
    if boolean_list_out is boolean_list_out_1:
        likelihood_values_1 = likelihood_values
    elif boolean_list_out is boolean_list_out_2:
        likelihood_values_2 = likelihood_values
    elif boolean_list_out is boolean_list_out_3:
        likelihood_values_3 = likelihood_values

plt.figure(figsize=(15, 10))

plt.subplot(3, 1, 1)
plt.plot(p_values, likelihood_values_1, label='MLE Dataset 1')
plt.xlabel('p (Probability of Head)')
plt.ylabel('Likelihood')
plt.title('Maximum Likelihood Function for Dataset 1')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(p_values, likelihood_values_2, label='MLE Dataset 2')
plt.xlabel('p (Probability of Head)')
plt.ylabel('Likelihood')
plt.title('Maximum Likelihood Function for Dataset 2')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(p_values, likelihood_values_3, label='MLE Dataset 3')
plt.xlabel('p (Probability of Head)')
plt.ylabel('Likelihood')
plt.title('Maximum Likelihood Function for Dataset 3')
plt.legend()

plt.tight_layout()

plt.show()
```

```python
# Problem 1 E

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

def bootstrap(dataset, sample_size, iterations=100):
    expectation_values = []
    
    for _ in range(iterations):
        bootstrap_sample = np.random.choice(dataset, size=sample_size, replace=True)
        expectation_values.append(np.mean(bootstrap_sample))
    
    return expectation_values


datasets = [boolean_list_out_1, boolean_list_out_2, boolean_list_out_3]

sample_sizes = [5, 15, 40, 60, 90, 150, 210, 300, 400]

bootstrap_results = []

for dataset in datasets:
    dataset_expectation_values = []
    for sample_size in sample_sizes:
        expectation_values = bootstrap(dataset, sample_size, iterations=100)
        dataset_expectation_values.append(expectation_values)
    bootstrap_results.append(dataset_expectation_values)

num_datasets = len(datasets)
num_sample_sizes = len(sample_sizes)

num_subplots = num_datasets * num_sample_sizes

fig, axes = plt.subplots(num_datasets, num_sample_sizes, figsize=(15, 15))

axes = axes.flatten()

for i, dataset in enumerate(datasets):
    for j, sample_size in enumerate(sample_sizes):
        ax = axes[i * num_sample_sizes + j]
        if j >= len(bootstrap_results[i]):
            break
        ax.hist(bootstrap_results[i][j], bins=15, edgecolor='k', alpha=0.7)
        ax.set_title(f'Dataset {i+1}\nSample Size {sample_size}')
        ax.set_xlabel('Expectation Value')
        ax.set_ylabel('Frequency')

plt.tight_layout()
plt.subplots_adjust(wspace=0.8, hspace=0.5)

plt.show()

for i, dataset_expectation_values in enumerate(bootstrap_results):
    for j, sample_size in enumerate(sample_sizes):
        if j >= len(dataset_expectation_values):
            break
        expectation = np.mean(dataset_expectation_values[j])
        variance = np.var(dataset_expectation_values[j])
        print(f'Dataset{i+1},SampleSize {sample_sizes[j]}:')
        print(f'Expectation Value: {expectation:.8f}')
        print(f'Variance: {variance:.8f}')

```

```python
# Problem 1 F
import numpy as np

datasets = [boolean_list_out_1, boolean_list_out_2, boolean_list_out_3]

sample_sizes = [5, 15, 40, 60, 90, 150, 210, 300, 400]

def bayesian_inference(dataset, sample_size):
    p = np.mean(dataset)
    N = len(dataset)
    M = p * N
    variance = (p * (1 - p)) / N
    return M, variance

def bootstrap(dataset, sample_size, iterations=100):
    expectation_values = []
    variances = []
    
    for _ in range(iterations):
        bootstrap_sample = np.random.choice(dataset, size=sample_size, replace=True)
        expectation_values.append(np.mean(bootstrap_sample))
        variances.append(np.var(bootstrap_sample))
    
    return expectation_values, variances

bayesian_expectation_values = []
bayesian_variances = []
bootstrap_expectation_values = []
bootstrap_variances = []

for dataset in datasets:
    bayesian_exp_values_dataset = []
    bayesian_var_dataset = []
    bootstrap_exp_values_dataset = []
    bootstrap_var_dataset = []
    
    for sample_size in sample_sizes:
        M, var = bayesian_inference(dataset, sample_size)
        bayesian_exp_values_dataset.append(M)
        bayesian_var_dataset.append(var)
        
        bootstrap_exp_values, bootstrap_var = bootstrap(dataset, sample_size, iterations=100)
        bootstrap_exp_values_dataset.append(np.mean(bootstrap_exp_values))
        bootstrap_var_dataset.append(np.mean(bootstrap_var))
    
    bayesian_expectation_values.append(bayesian_exp_values_dataset)
    bayesian_variances.append(bayesian_var_dataset)
    bootstrap_expectation_values.append(bootstrap_exp_values_dataset)
    bootstrap_variances.append(bootstrap_var_dataset)
    
    import matplotlib.pyplot as plt


fig, axes = plt.subplots(2, 3, figsize=(18, 12))

axes = axes.flatten()

plot_labels = ['Dataset 1', 'Dataset 2', 'Dataset 3']

for i in range(3):  
    ax = axes[i]  # First row
    ax.plot(sample_sizes, bayesian_expectation_values[i], label='Bayesian Inference', marker='o')
    ax.plot(sample_sizes, bootstrap_expectation_values[i], label='Bootstrapping', marker='o')
    ax.set_xlabel('Sample Size')
    ax.set_ylabel('Expectation Value')
    ax.set_title(f'Expectation Value Comparison - {plot_labels[i]}')
    ax.legend()

for i in range(3):  
    ax = axes[i + 3]  # Second row
    ax.plot(sample_sizes, bayesian_variances[i], label='Bayesian Inference', marker='o')
    ax.plot(sample_sizes, bootstrap_variances[i], label='Bootstrapping', marker='o')
    ax.set_xlabel('Sample Size')
    ax.set_ylabel('Variance')
    ax.set_title(f'Variance Comparison - {plot_labels[i]}')
    ax.legend()

plt.tight_layout()
plt.show()
```
